### Calculate centroid data on all levels

```python
def convert_coordinates_format(geometry):
    """
    Convert coordinates from GeoDataFrame geometry (tuple format) to the required format.
    
    Parameters:
        geometry (shapely.geometry): The geometry object (Polygon or MultiPolygon).
        
    Returns:
        list: Coordinates in the required format, i.e., a list of lists of [longitude, latitude].
    """
    if geometry.geom_type == 'MultiPolygon':
        # convert each Polygon within the MultiPolygon to the required format
        return [
            [list(polygon.exterior.coords) for polygon in geometry.geoms]
        ]
    elif geometry.geom_type == 'Polygon':
        # convert the Polygon to the required format
        return [list(geometry.exterior.coords)]
    else:
        return []
```

```python
gdf_admin_all["COORDINATES_L1"] = gdf_admin_all["GEOMETRY_L1"].progress_apply(lambda x: convert_coordinates_format(x))
gdf_admin_all["COORDINATES_L2"] = gdf_admin_all["GEOMETRY_L2"].progress_apply(lambda x: convert_coordinates_format(x))
gdf_admin_all["COORDINATES_L3"] = gdf_admin_all["GEOMETRY_L3"].progress_apply(lambda x: convert_coordinates_format(x))
```

```python
def get_centroid_coordinates(gdf, level):
    """
    Get the centroid of each region in the GeoDataFrame.
    
    Args:
    - gdf (GeoDataFrame): A GeoDataFrame with polygon or multipolygon geometries.
    
    Returns:
    - list: A list of (latitude, longitude) tuples for each region's centroid.
    """
    centroids = []
    
    for _, row in gdf.iterrows():
        geometry = row[f'GEOMETRY_L{level}']
        
        if geometry.is_valid:
            centroid = geometry.centroid
            centroids.append((centroid.y, centroid.x))  # (lat, lon)
    
    return centroids
```

```python
gdf_admin_all["CENTROID_COORDINATES_L1"] = get_centroid_coordinates(gdf_admin_all, "1")
gdf_admin_all["CENTROID_COORDINATES_L2"] = get_centroid_coordinates(gdf_admin_all, "2")
gdf_admin_all["CENTROID_COORDINATES_L3"] = get_centroid_coordinates(gdf_admin_all, "3")
```

```python
gdf_admin_all.to_csv('data/merging/gadm_sdn_full_with_coords_and_centroids.csv', index=False)
```

```python
gdf_weather = gdf_weather[['GID_1', 'NAME_1', 'date', 'cloud_cover',
       'humidity_percentage', 'precipitation_total_mm',
       'temperature_min_celsius', 'temperature_max_celsius',
       'temperature_afternoon_celsius', 'temperature_night_celsius',
       'temperature_evening_celsius', 'temperature_morning_celsius',
       'pressure_hpa', 'wind_max_speed_kmh', 'wind_max_direction_degrees']].copy().reset_index(drop=True)
```

```python
gdf_weather.columns = ['GID_L1', 'NAME_L1', 'date', 'cloud_cover',
       'humidity_percentage', 'precipitation_total_mm',
       'temperature_min_celsius', 'temperature_max_celsius',
       'temperature_afternoon_celsius', 'temperature_night_celsius',
       'temperature_evening_celsius', 'temperature_morning_celsius',
       'pressure_hpa', 'wind_max_speed_kmh', 'wind_max_direction_degrees']
```

```python
def aggregate_weather(group):
    return group.apply(
        lambda row: {
            row["date"].strftime("%Y-%m-%d"): {
                "cloud_cover": row["cloud_cover"],
                "humidity_percentage": row["humidity_percentage"],
                "precipitation_total_mm": row["precipitation_total_mm"],
                "temperature_min_celsius": row["temperature_min_celsius"],
                "temperature_max_celsius": row["temperature_max_celsius"],
                "temperature_afternoon_celsius": row["temperature_afternoon_celsius"],
                "temperature_night_celsius": row["temperature_night_celsius"],
                "temperature_evening_celsius": row["temperature_evening_celsius"],
                "temperature_morning_celsius": row["temperature_morning_celsius"],
                "pressure_hpa": row["pressure_hpa"],
                "wind_max_speed_kmh": row["wind_max_speed_kmh"],
                "wind_max_direction_degrees": row["wind_max_direction_degrees"],
            }
        },
        axis=1,
    ).tolist()
```

```python
gdf_weather_gr = gdf_weather.groupby(['GID_L1', 'NAME_L1']).apply(aggregate_weather).reset_index(name="WEATHER_WEEKLY_L1")
```

```python
gdf_weather_gr["WEATHER_WEEKLY_L1"] = gdf_weather_gr["WEATHER_WEEKLY_L1"].apply(lambda x: {k: v for d in x for k, v in d.items()})
```

```python
gdf_admin_sentinel_weather = pd.merge(gdf_admin_sentinel, gdf_weather_gr, on=['GID_L1', 'NAME_L1'], how='left')
```
